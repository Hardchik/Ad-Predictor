```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
#Import Dataset
data = pd.read_csv('/content/Ads_CTR_Optimisation.csv')
data
```

```python
#Implement Thompson Sampling
N = 10000
ad = 10
adSelected = []
rewardNumber1 = [0] * ad
rewardNumber0 = [0] * ad
totalReward = 0

for n in range(0, N):
  AD = 0
  maxRandom = 0
  for i in range(0 ,ad):
    randomBeta = random.betavariate(rewardNumber1[i]+1, rewardNumber0[i]+1)

    if randomBeta > maxRandom:
      maxRandom = randomBeta
      AD = i

  adSelected.append(AD)
  reward = data.values[n, AD]
  if reward == 1:
    rewardNumber1[AD] += 1
  else :
    rewardNumber0[AD] += 1
      
  totalReward += reward
```

```python
#visualize the result
plt.hist(adSelected)
plt.title('Visualizing the results')
plt.xlabel('Ads')
plt.ylabel('Number of selection of Ads')
```
